## ***5. ¬øQu√© es un decorador de python?***

@ [decoradores ‚ñ∂](https://youtu.be/z-5cQA9EKdw)

Los decoradores son una potente caracter√≠stica de Python que permite modificar o mejorar funciones, clases u otros objetos en tiempo de ejecuci√≥n sin modificar directamente su c√≥digo

Los decoradores se implementan utilizando funciones o clases y se utilizan para envolver o decorar el objeto de destino con funcionalidad adicional.

Ventajas:
* Reutilizaci√≥n del c√≥digo
* Facilidad de lectura y mantenimiento
* Extensibilidad :
     Proporcionan un mecanismo flexible para ampliar y personalizar el comportamiento de funciones o clases sin modificar su definici√≥n original.
* Los decoradores pueden aceptar argumentos, lo que permite personalizar el comportamiento de la funci√≥n decorada
* Encadenar varios


SINTAXIS:

Los decoradores se implementan utilizando el s√≠mbolo **"@"** seguido del nombre de la funci√≥n o clase decoradora. Se colocan antes de la definici√≥n de la funci√≥n que se va a decorar. La funci√≥n/clase decoradora se encarga de modificar o mejorar el comportamiento de la funci√≥n de destino. He aqu√≠ un ejemplo:
````
@decorator
def my_function():
    # Function code
````

`````python
# Ejemplo:

# Tenemos una funci√≥n suma() que vamos a decorar usando mi_decorador(). 
# Para ello, antes de la declaraci√≥n de la funci√≥n suma, hacemos uso de @mi_decorador.

def mi_decorador(funcion):
    def nueva_funcion(a, b):
        print("Se va a llamar")
        c = funcion(a, b)
        print("Se ha llamado")
        return c
    return nueva_funcion

@mi_decorador
def suma(a, b):
    print("Entra en funcion suma")
    return a + b

suma(5,8)

# Se va a llamar
# Entra en funcion suma
# Se ha llamado
`````
